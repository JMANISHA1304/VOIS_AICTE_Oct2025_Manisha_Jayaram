```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
#Read the AirBnb Data
df = pd.read_excel('/content/1730285881-Airbnb_Open_Data.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
df.duplicated().value_counts()
```

```python
df[df['neighbourhood group']=='brookln']
```

```python
#Drop duplicate records
df.drop_duplicates(inplace=True)
```

```python
df.drop(['house_rules','license'], axis=1, inplace = True)
```

```python
df['price'] = df['price'].astype(str).str.replace('$', '', regex=False)
```

```python
df['service fee'] = df['service fee'].astype(str).str.replace('$', '', regex=False)
```

```python
df['price'] = df['price'].astype(str).str.replace(',', '', regex=False)
```

```python
df['service fee'] = df['service fee'].astype(str).str.replace(',', '', regex=False)
```

```python
df.rename(columns={
    'price':'price_$',
    'service fee':'service_fee_$'
}, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
#change all mismatched data types to the appropriate once
df['price_$'] = df['price_$'].astype(float)
df['service_fee_$'] = df['service_fee_$'].astype(float)
df['id'] = df['id'].astype(str)
df['host id'] = df['host id'].astype(str)
df['last review'] = pd.to_datetime(df['last review'])
df['Construction year'] = df['Construction year'].astype(int)
```

```python
#correct the spelling of 'brookln' to brooklyn
df.loc[df['neighbourhood group'] == 'brookln', 'neighbourhood group'] = 'Brooklyn'
```

```python
#get rid of outliers in 'availability 365' column data
df = df.drop(df[df['availability 365'] > 365].index)
```

EDA(EXPLORATORY DATA ANALYSIS)

```python
df.describe()
```

```python
property_types = df['room type'].value_counts().to_frame()
property_types
```

```python
room_type_bar = plt.bar(property_types.index, property_types.loc[:,'count']);
plt.bar_label(room_type_bar, labels = property_types.loc[:,"count"], padding=4);
plt.ylim([0,50000])
plt.title('Property Types and their count in the Dataset')
plt.xlabel('Room Type')
plt.ylabel('Room Type Count')
plt.show()
```

Which Neighbourhood group has the highest number of listings?

```python
hood_group = df['neighbourhood group'].value_counts().to_frame()
hood_group
```

```python
hood_group_bar = plt.bar(hood_group.index, hood_group.loc[:,'count']);
plt.bar_label(hood_group_bar, labels = hood_group.loc[:,"count"], padding=4);
plt.ylim([0,50000])
plt.xlabel('Neighbourhood Groups')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45);
plt.title('Which Neighbourhood Groups has the heighest number of Listings')
plt.show()
```

Which neighbourhood group have the highest average prices for Airbnb listings?

```python
avg_price = df.groupby('neighbourhood group')['price_$'].mean().sort_values(ascending=False).to_frame()
avg_price_bar = plt.bar(avg_price.index, avg_price.loc[:,'price_$']);
plt.bar_label(avg_price_bar, labels = round(avg_price.loc[:,"price_$"], 2),label_type='edge', padding=4);
plt.ylim([0,1000])
plt.xlabel('Neighbourhood Groups')
plt.ylabel('Average Price Per Listing($)')
plt.xticks(rotation=45);
plt.title('Average Price Per Listings ($) in each Neighbour Group')
plt.show()
```

Is there a relationship between the construction year of property and price?

```python
#Is there a relationship between the construction year of property and price?
df.groupby(df['Construction year'])['price_$'].mean().to_frame().plot()
plt.xlabel('Construction Year')
plt.ylabel('Average Price($)')
plt.title('Relationship between Construction Year and Price')
plt.show()
```

who are the top 10 hots by calculated host listing count?

```python
#who are the top 10 hots by calculated host listing count?
top_10_hosts = df.groupby('host name')['calculated host listings count'].sum().sort_values(ascending=False).head(10).to_frame()
top_10_hosts_bar = plt.bar(top_10_hosts.index, top_10_hosts.loc[:,'calculated host listings count']);
plt.bar_label(top_10_hosts_bar, labels = round(top_10_hosts.loc[:,"calculated host listings count"], 2),label_type='edge', padding=4);
plt.xlabel('Hosts Name')
plt.ylabel('Calculated host of Listings')
plt.xticks(rotation=80);
plt.ylim([0,140000])
plt.title('Top 10 Hosts by Calculated Host Listings Count')
plt.show()
```

Are hosts with verified identitites more likely to recieve positive review

```python
#Are hosts with verified identitites more likely to recieve positive review
review = df.groupby('host_identity_verified')['review rate number'].mean().sort_values(ascending=False).to_frame()
review
```

```python
review_bar = plt.bar(review.index, review.loc[:,'review rate number']);
plt.bar_label(review_bar, labels = round(review.loc[:,"review rate number"], 2), padding=4);
plt.ylim([0,4])
plt.xlabel('Host Identity Verified')
plt.ylabel('Average Review Rate')
plt.title('Average Review Rate for each verification statistics')
plt.show()
```

```python
base_color = sns.color_palette()[0]
sns.boxplot(data = df, x="host_identity_verified", y="review rate number", color = base_color)
plt.xlabel('Host Verification status')
plt.ylabel('Review Rate Number')
plt.title('Average Review Rate Number for each verification status')
plt.show()
```
